# TUSZ Example
This notebook gives an example of how to use the package with the TUSZ database.

```python
import seizure_data_processing as sdp
import numpy as np
from seizure_data_processing.config import TUSZ_DIR
```

```python
# Load a single file
file = (
    TUSZ_DIR
    + r"/edf/train/01_tcp_ar/027/00002707/s001_2006_03_17/00002707_s001_t001.edf"
)
channels = "FP1;FP2;F3;F4;C3;C4;P3;P4;O1;O2;F7;F8;T3;T4;T5;T6;CZ".split(";")
eeg_file = sdp.EEG(file, channels=channels)
```
